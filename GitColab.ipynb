**Installing on Windows**

There are also a few ways to install Git on Windows. The most official build is available for download on the Git website. Note that this is a project called Git for Windows, which is separate from Git itself; for more information on it, go to https://gitforwindows.org.

To get an automated installation you can use the Git Chocolatey package. Note that the Chocolatey package is community maintained.
